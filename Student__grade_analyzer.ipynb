```python
def calculate_grades():
    print("Student Grade Analyzer")

    marks = []
    n = int(input("Enter number of subjects: "))

    for i in range(n):
        m = float(input(f"Enter marks for subject {i+1}: "))
        marks.append(m)

    average = sum(marks) / n
    highest = max(marks)
    lowest = min(marks)

    print("\n--- Result ---")
    print("Average Marks:", round(average, 2))
    print("Highest Marks:", highest)
    print("Lowest Marks:", lowest)

    if average >= 90:
        print("Grade: A+")
    elif average >= 75:
        print("Grade: A")
    elif average >= 60:
        print("Grade: B")
    elif average >= 50:
        print("Grade: C")
    else:
        print("Grade: Fail")

calculate_grades()
```
